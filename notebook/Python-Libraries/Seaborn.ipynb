```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap
A correlation heatmpa uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table)
between two discrete dimensions or event types. It is very important in Feature Selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

## JoinPlot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex');
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg');
```

## PairPlot

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue='sex')
```

## DistPlot

```python
sns.distplot(df['tip']);
```

```python
sns.distplot(df['tip'], kde=False, bins=10);
```

## Categorical Plots
* Boxplot
* violinplot
* countplot
* barplot

## Countplot

```python
sns.countplot('sex', data=df);
```

```python
sns.barplot(y='total_bill', x='smoker', data=df);
```

```python
sns.barplot(y='total_bill', x='sex', data=df);
```

```python
sns.boxplot(y='total_bill', x='smoker', data=df);
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df);
```

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow');
```
