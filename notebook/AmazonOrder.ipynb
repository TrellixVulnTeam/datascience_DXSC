**Replacing the values of order_status to 1 and 0**

**1-Delivered to buyer**

**0-Returned to seller**

```python
# Making a copy of DataFrame
df_copy = df.copy()
```

```python
df_copy["order_status"] = df_copy["order_status"].map({"Delivered to buyer":1,
                            "Returned to seller":0})
```

```python
df_copy
```

## Using the Machine Learning Model to predict how many products are getting ReturnedToSeller

**Using LinearSVC**

```python
# Importing The estimator
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

# Setting up the random seed
np.random.seed(42)

# Making the data
X = df_copy.drop("order_status", axis=1)
y = df_copy["order_status"]

# Split the data into training and test model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiating LinearSVC
clf = LinearSVC()
clf.fit(X_train, y_train)

# Scoring the model
clf.score(X_test, y_test)
```
