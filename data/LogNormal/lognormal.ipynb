```python
import pandas as pd
import numpy as np
```

## First_Bet

```python
first_bet = pd.read_excel("./First_Bet_Data.xlsx")
first_bet.head(10)
```

```python
missing_val = [features for features in first_bet.columns if first_bet[features].isnull().sum()>1]

for feature in missing_val:
    print(feature, np.round(first_bet[feature].isnull().mean(), 3), "% missing values")
```

```python
first_bet.dtypes
```

```python
for i in first_bet.columns:
    if first_bet[i].dtypes == 'object':
        first_bet[i].fillna(first_bet[i].mode()[0], inplace=True)
    else:
        first_bet[i].fillna(first_bet[i].median(), inplace=True)
print(first_bet)
```

```python
first_bet.head(10)
```

```python
first_bet.to_excel("first_bet.xlsx", index=False)
```
